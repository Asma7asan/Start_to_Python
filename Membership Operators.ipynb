# Type  Of Python Operators:
# 6 - Membership Operators

```python
# in : returns True if the value exists in the sequence.
list_1 = [1, 2, 3, 4]

print(2 in list_1)   # True
print(5 in list_1)   # False
```

```python
name = "python"
print("o" in name)   # True
print("z" in name)   # False
```

```python
# not in : returns True if the value does not exist in the sequence.
list_1 = [1, 2, 3, 4]

print(5 not in list_1)   # True
print(2 not in list_1)   # False
```

```python
text = "python"
print("z" not in text)   # True
print("y" not in text)   # False
```
