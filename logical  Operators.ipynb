# Type  Of Python Operators:
# 4 - logical  Operators

```python
# and :Returns True only if both conditions are True
num_1 = 12
num_2 = 45
print( num_1 > 10 and num_2 < 50)
print( num_1 > 10 and num_2 > 50)
```

```python
# or : Returns True if at least one condition is True.
num_1 = 12
num_2 = 45
print( num_1 > 10 or num_2 < 50)
print( num_1 > 10 or num_2 > 50)
```

```python
# not : Reverses the result. True → False | | False → True
value = False
print( value )
print(not value)
```

```python
x = 10
print(x > 5 and x < 20)   # True
print(x > 15 or x == 10)  # True
print(not(x == 10))  
```
