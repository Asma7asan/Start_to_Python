# A for loop is used to repeat a block of code a specific number of times or to iterate over items in a sequence (like a list, string, tuple, etc.).

```python
# Loop with range
for i in range(5):
    print(i)
```

```python
# Loop with start and end in range
for i in range(2, 6):
    print(i)
```

```python
# Loop through a list
language = ["Python", "Java", "Html","CSS"]
for L in language:
    print(L)
```

```python
# Loop with step
for i in range(0, 10, 2):
    print(i)
```

```python
# Loop through characters in a string
for char in "Hello":
    print(char)
```

# Example

```python
# Print numbers from 100 to 110.
for i in range(100, 111):
    print(i)
```

```python
# Sum of numbers
total = 0
for i in range(1, 51):
    total += i
print(total)
```

```python
# Count vowels
text = "I love programming"
vowels = "aeiouAEIOU"
count = 0

for ch in text:
    if ch in vowels:
        count += 1

print(count)
```

```python
# Find the maximum number (without max())
numbers = [4, 99, 12, 7, 34]
max_num = numbers[0]

for n in numbers:
    if n > max_num:
        max_num = n

print(max_num)
```

```python
# Multiplication table (5)
for i in range(1, 11):
    print("5 x", i, "=", 5 * i)
```

```python
# Remove duplicates
items = [1, 2, 2, 3, 4, 4, 5]
unique = []

for item in items:
    if item not in unique:
        unique.append(item)

print(unique)
```

```python
# Reverse a string
text = "Hello"
rev = ""

for ch in text:
    rev = ch + rev   # add characters to the front

print(rev)
```

```python
# Count even and odd numbers
nums = [12, 45, 66, 23, 90, 11]
even = 0
odd = 0

for n in nums:
    if n % 2 == 0:
        even += 1
    else:
        odd += 1

print("Even:", even)
print("Odd:", odd)
```

```python
# Print a triangle
for i in range(1, 5):
    print("*" * i)
```

```python
# Print a reversed triangle 
for i in range(4,0,-1):
    print("*" * i)
```

```python
# Print a Pyramid
rows = 5
for i in range(1, rows + 1):
    print(" " * (rows - i) + "*" * (2 * i - 1))
```

```python
# Inverted Pyramid
rows = 5
for i in range(rows, 0, -1):
    print(" " * (rows - i) + "*" * (2 * i - 1))
```

```python
# Hollow Square
n = 5
for i in range(n):
    if i == 0 or i == n-1:
        print("*" * n)
    else:
        print("*" + " " * (n-2) + "*")
```
