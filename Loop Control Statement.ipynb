# Loop Control Statements change the normal flow of loops. They allow to skip an iteration, stop the loop early, or move to another loop level.
# There are three main loop control statements:
# 1 - break : Terminate the loop statement and transfers execution to the statement immediately following the loop.
# 2 - continue :  causes the loop to skip the remainder of its body and immediately retest its condition prior to reiterating
# 3 - Pass : Used as a placeholder to avoid errors when the code block is required but you don’t want to write code yet.

```python
# break
for i in range(1, 10):
    if i == 5:
        break
    print(i)
```

```python
# continue
for i in range(1, 6):
    if i == 3:
        continue
    print(i)
```

```python
# Pass : (pass does nothing)
for i in range(5):
    if i == 2:
        pass
    print(i)
```

```python
# loop control
sentens = "loop con"
print(" 1 - Break :")
for letter in sentens:
    if letter == "p":
        break
    print(letter)
print(" 2 -continu :")
for letter in sentens:
    if letter == "p":
        continue
    print(letter)
print(" 3 - pass :")
for letter in sentens:
    if letter == "p":
        pass
    print(letter)
```
