# Arithmetic Operators


```python
# Addition
print(20 + 30)
print(-10 + 30)
print(10 + 2.66)
print(1.2 + 7.2)
```

```python
# Subtraction
print(60 - 40) 
print(-30 - 10)
print(-30 - -20)  
print(4.66 - 3.44)
```

```python
# Multiplication
print( 10 * 30)
print( 5 + 10 * 100)
print( (5 + 10) * 100)
```

```python
# Division
print(200 / 20)
print(int(200 / 20))
print(float( 12 / 4 ))
```

```python
# Modulus
print(8 % 2)
print(9 % 2)
print(20 % 5)
print(22 % 5)
```

```python
# Exponent
print(2 ** 5)
print(2 * 2 * 2 * 2 * 2)
print(5 ** 2)
print(5 * 5 * 5 * 5 )
```

```python
# Floor Division
print(100 // 20)
print(119 // 20)
print(120 // 20)
print(140 // 20)
print(142 // 20)

```
