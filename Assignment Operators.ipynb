# Assignment Operators

```python
# Assignment Operator
a = 10 
b = 20 
print(a)   
print(b)
```

```python
# Addition Assignment
a = 10
print(a) 
a += 4              # Same as a = 10 + 4 
print(a) 
```

```python
# Subtraction Assignment
a = 10
print(a) 
a -= 4              # Same as a = 10 - 4 
print(a) 
```

```python
# Multiplication Assignment  
a = 10
print(a) 
a *= 4              # Same as a = 10 * 4 
print(a) 
```

```python
# Division Assignment  
a = 10
print(a) 
a /= 4              # Same as a = 10 / 4 
print(a) 
```

```python
# Exponent Assignment 
a = 10
print(a) 
a **= 4              # Same as a = 10 ** 4 
print(a) 
```

```python
# Remainder Assignment           
a = 10
print(a) 
a //= 4              # Same as a = 10 // 4 
print(a)
```
