# Type  Of Python Operators:
# 5 - Bitwise  Operators

```python
# Binary And : set each bit to 1 if both bits are 1
num_1 = 0b0111100
num_2 = 0b0001101
print(num_1 & num_2)
```

```python
# Binary or : set each bit to 1 if one of two bits is 1
num_1 = 0b0111100
num_2 = 0b0001101
print(num_1 | num_2)
```

```python
# Binary xor : set each bit to 1 if only one of two bits is 1
num_1 = 0b0111100
num_2 = 0b0001101
print(num_1 ^ num_2)
```

```python
# Binary ones complement : inverts all the bits
num_1 = 0b0111100
num_2 = 0b0001101
print(~num_1)
print(~num_2)
```

```python
# Binary left shift : shift left by pushing zeros in from the right and let the leftmost bits fall off
num_1 = 0b0111100
num_2 = 0b0001101
print(num_1<<2)
print(num_2<<2)
```

```python
# Binaryright shift : shift right by pushing copies of the leftmost bit in from the left and let the rightmost bits fall off
num_1 = 0b0111100
num_2 = 0b0001101
print(num_1>>2)
print(num_2>>2)
```
