# Comparison Operators
#      *All of these operations return a Boolean value True or false

```python
# Equal
print(100 == 100)
print(100 == 200)
print(100 == 100.00)
```

```python
# Not Equal
print(100 != 100)
print(100 != 200)
print(100 != 100.00) 
```

```python
# Greater Than 
print(100 > 100)
print(100 > 200)
print(100 > 100.00)
print(100 > 40)
```

```python
# Less Than
print(100 < 100)
print(100 < 200)
print(100 < 100.00)
print(100 < 40)
```

```python
#  Greater Than Or Equal
print(100 >= 100)
print(100 >= 200)
print(100 >= 100.00)
print(100 >= 40)
```

```python
# Less Than Or Equal
print(100 <= 100)
print(100 <= 200)
print(100 <= 100.00)
print(100 <= 40)
```
