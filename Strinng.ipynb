```python
# Strinng opreation
word = "Hello in Python"
print(word[0])
print(word[7:9])
print(word[-6:-1])
print(word[::2])
print(word[::-1])
print(word[2:])
print(word * 3)
print(word + " - Enjoy Coding!")
print(word[1:10:2])
```

```python
# Basic data types and operationss of variables
number = 1234
print(type(number))
string = "1234"
print(type(string))
float_number = 12.34
print(type(float_number))
boolean_value = True
print(type(boolean_value))  
complex_number = 1 + 2j
print(type(complex_number)) 
```

```python
# upper and lower case conversion
text = "Hello World"
print(text.upper())
print(text.lower())
```

```python
# String formatting
name = "Alice"
age = 30
formatted_string = f"My name is {name} and I am {age} years old."
print(formatted_string)
```

```python
# Capitalize the first letter of each word
word = "hello in python"
print(word.capitalize())
```

```python
# Count occurrences of a substring
sentence = "This is a test. This test is in python."
count_test = sentence.count("test")
print(f"The word 'test' occurs {count_test} times.")   # using f-string for formatting
```

```python
# Split : Splits a string into a list based on a specified delimiter.
sentence = "This is a test in python."
print(sentence.split(" "))
print(sentence.split("is"))
word = "Hello World in Python"
print(word.split('o' , 1))    # Split only at the first occurrence of 'o'
print(word.split('o' , 3))    # Split at the first three occurrences of 'o'
print(word.rsplit('o' , 1))   # Split from the right at the first occurrence of 'o'
```

```python
# len : Returns the length of the string.
sentence = "Hello, welcome to Python programming."
length = len(sentence)
print(length) 
```

```python
# strip : Removes leading and trailing whitespace (or specified characters) from the string.
sentence = "      Hello,  How are you Engineer?       "
print(sentence.strip())          # Removes leading and trailing whitespace
print(sentence.lstrip())        # Removes leading whitespace
print(sentence.rstrip())        # Removes trailing whitespace
print(sentence.strip(" "))      # Removes leading and trailing spaces
print(sentence.strip(" ?"))    # Removes leading and trailing spaces and question marks
```

```python
# title : Converts the first character of each word to uppercase and the rest to lowercase.
sentence = "hello every one! welcome to python programming."
print(sentence.title())
```

```python
# zfill : Pads the string on the left with zeros to fill a specified width.
number_str = "42"
padded_number = number_str.zfill(5)
print(padded_number)  # Output: "00042" 
```

```python
# Center : Centers the string within a specified width, padding with spaces or a specified character.
word = "Hello"
centered_word = word.center(20, '*')
print(centered_word)  # Output: "*******Hello********"
centered_word = word.center(20, '.')
print(centered_word)  # Output: " .  .  . Hello .  .  . "
```

```python
# swapcase : Swaps the case of each character in the string.
sentence = "Hi I am Engineer"
print(sentence.swapcase())  # Output: "hI i AM eNGINEER"
```

```python
# startswith : Checks if the string starts with a specified prefix.
sentence = "Hello, welcome to Python programming."
print(sentence.startswith("Hello"))  # Output: True
print(sentence.startswith("welcome"))  # Output: False
```

```python
# endwith : Checks if the string ends with a specified suffix.
sentence = "Hello, welcome to Python programming."
print(sentence.endswith("programming."))  # Output: True
print(sentence.endswith("Python"))  # Output: False
```

```python
# index : Finds the first occurrence of a substring and returns its index.
sentence = "This is a test in python. This test is fun."
print(sentence.index("test"))  # Output: 10
print(sentence.index("is", 20 , 38))  # Output: 5 (search starts from index 20 end at index 38)
```

```python
# find : Finds the first occurrence of a substring and returns its index, or -1 if not found.
sentence = "This is a test in python. This test is fun."
print(sentence.find("python"))  # Output: 16
```

```python
# rjust : Right-justifies the string within a specified width, padding with spaces or a specified character.
word = "Hello"
right_justified_word = word.rjust(20, '#')
print(right_justified_word)  # Output: "###############Hello"
right_justified_word = word.rjust(40, '>')
print(right_justified_word)  # Output: "<><><><><><><><><>Hello"

```

```python
# ljust : Left-justifies the string within a specified width, padding with spaces or a specified character.
word = "Hello"
left_justified_word = word.ljust(20, '-')
print(left_justified_word)  # Output: "Hello---------------"
right_justified_word = word.ljust(20, '>')
print(right_justified_word)  # Output: "Hello>>>>>>>>>>>>>>>"
```

```python
# splitline : Splits a string into a list at line breaks.
multiline_string = "Hello World\nWelcome to Python programming.\nEnjoy coding!" 
print(multiline_string.splitlines())
```

```python
# expandtabs : Replaces tab characters with spaces, using a specified tab size.
multiline_string = "Hello\tWorld!\nWelcome\tto\tPython\tprogramming."
print(multiline_string.expandtabs(tabsize=4))
```

```python
# replace : Replaces occurrences of a specified substring with another substring.
sentence = "The quick brown fox jumps over the lazy dog. The fox is clever."
new_sentence = sentence.replace("fox", "cat")
print(new_sentence)  # Output: "The quick brown cat jumps over the lazy dog. The cat is clever."
```

```python
# is 
word = "HelloWorld"
number = 12
float_number = 12.34
boolean_value = True
print(word.isalpha())        # True, all characters are alphabetic
print(str(number).isnumeric())   # True, all characters are numeric 
print(str(float_number).isdecimal())  # False, contains a decimal point
print(str(boolean_value).isidentifier())  # True, 'True' is a valid identifier
print(word.islower())      # False, not all characters are lowercase
print(word.isupper())      # False, not all characters are uppercase
print(word.isspace())      # False, contains non-whitespace characters  
print("   ".isspace())     # True, all characters are whitespace
```
