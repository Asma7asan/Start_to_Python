# Python data type conversion

```python
# int to str conversion
num = 42
num_str = str(num)
print(f"Integer: {num}, String: '{num_str}'")   
# int to float conversion
num_float = float(num)
print(f"Integer: {num}, Float: {num_float}")
# int to complex conversion
num_complex = complex(num)
print(f"Integer: {num}, Complex: {num_complex}")
# int to bool conversion
num_bool = bool(num)
print(f"Integer: {num}, Boolean: {num_bool}")
# int to set conversion
num_set = set([num])
print(f"Integer: {num}, Set: {num_set}")
# int to list conversion
num_list = [num]
print(f"Integer: {num}, List: {num_list}")
# int to tuple conversion
num_tuple = (num,)
print(f"Integer: {num}, Tuple: {num_tuple}")
# int to dict conversion
num_dict = { 'number': num }
print(f"Integer: {num}, Dictionary: {num_dict}")  
```

```python
# float to str conversion
flt = 3.14
flt_str = str(flt)
print(f"Float: {flt}, String: '{flt_str}'")
# float to int conversion
flt_int = int(flt)
print(f"Float: {flt}, Integer: {flt_int}")
# float to complex conversion
flt_complex = complex(flt)
print(f"Float: {flt}, Complex: {flt_complex}")
# float to bool conversion
flt_bool = bool(flt)
print(f"Float: {flt}, Boolean: {flt_bool}")
# float to set conversion
flt_set = set([flt])
print(f"Float: {flt}, Set: {flt_set}")
# float to list conversion
flt_list = [flt]
print(f"Float: {flt}, List: {flt_list}")
# float to tuple conversion
flt_tuple = (flt,)
print(f"Float: {flt}, Tuple: {flt_tuple}")
# float to dict conversion
flt_dict = { 'float': flt }
print(f"Float: {flt}, Dictionary: {flt_dict}")
```

```python
# str to int conversion
s = "123"
s_int = int(s)
print(f"String: '{s}', Integer: {s_int}")
# str to float conversion
s_float = float(s)
print(f"String: '{s}', Float: {s_float}")
# str to complex conversion
s_complex = complex(s)
print(f"String: '{s}', Complex: {s_complex}")
# str to bool conversion
s_bool = bool(s)
print(f"String: '{s}', Boolean: {s_bool}")
# str to set conversion
s_set = set(s)
print(f"String: '{s}', Set: {s_set}")
# str to list conversion
s_list = list(s)
print(f"String: '{s}', List: {s_list}")
# str to tuple conversion
s_tuple = tuple(s)
print(f"String: '{s}', Tuple: {s_tuple}")
# str to dict conversion (using characters as keys with dummy values)
s_dict = { char: None for char in s }
print(f"String: '{s}', Dictionary: {s_dict}")
```

```python
# bool to int conversion
b = True
b_int = int(b)
print(f"Boolean: {b}, Integer: {b_int}")    
# bool to float conversion
b_float = float(b)
print(f"Boolean: {b}, Float: {b_float}")
# bool to str conversion
b_str = str(b)
print(f"Boolean: {b}, String: '{b_str}'")
# bool to complex conversion
b_complex = complex(b)
print(f"Boolean: {b}, Complex: {b_complex}")
# bool to set conversion
b_set = set([b])
print(f"Boolean: {b}, Set: {b_set}")
# bool to list conversion
b_list = [b]
print(f"Boolean: {b}, List: {b_list}")
# bool to tuple conversion
b_tuple = (b,)
print(f"Boolean: {b}, Tuple: {b_tuple}")
# bool to dict conversion
b_dict = { 'boolean': b }
print(f"Boolean: {b}, Dictionary: {b_dict}")
```

```python
# list to str conversion
lst = [1, 2, 3]
lst_str = str(lst)
print(f"List: {lst}, String: '{lst_str}'")
# list to int conversion (will raise an error)
# lst_int = int(lst)  # Uncommenting this line will raise a TypeError
# list to float conversion (will raise an error)
# lst_float = float(lst)  # Uncommenting this line will raise a TypeError
# list to bool conversion
lst_bool = bool(lst)
print(f"List: {lst}, Boolean: {lst_bool}")
# list to set conversion
lst_set = set(lst)
print(f"List: {lst}, Set: {lst_set}")
# list to tuple conversion
lst_tuple = tuple(lst)
print(f"List: {lst}, Tuple: {lst_tuple}")
# list to dict conversion (using indices as keys)
lst_dict = { i: lst[i] for i in range(len(lst)) }
print(f"List: {lst}, Dictionary: {lst_dict}")
```

```python
# dict to str conversion
d = {'a': 1, 'b': 2}
d_str = str(d)
print(f"Dictionary: {d}, String: '{d_str}'")
# dict to int conversion (will raise an error)
# d_int = int(d)  # Uncommenting this line will raise a TypeError
# dict to float conversion (will raise an error)
# d_float = float(d)  # Uncommenting this line will raise a TypeError
# dict to bool conversion
d_bool = bool(d)
print(f"Dictionary: {d}, Boolean: {d_bool}")
# dict to set conversion (using keys)
d_set = set(d.keys())
print(f"Dictionary: {d}, Set: {d_set}")
# dict to list conversion (using keys)
d_list = list(d.keys())
print(f"Dictionary: {d}, List: {d_list}")
# dict to tuple conversion (using keys)
d_tuple = tuple(d.keys())
print(f"Dictionary: {d}, Tuple: {d_tuple}")
```

```python
# tuple to str conversion
tup = (4, 5, 6)
tup_str = str(tup)
print(f"Tuple: {tup}, String: '{tup_str}'")
# tuple to int conversion (will raise an error)
# tup_int = int(tup)  # Uncommenting this line will raise a TypeError
# tuple to float conversion (will raise an error)
# tup_float = float(tup)  # Uncommenting this line will raise a TypeError
# tuple to bool conversion
tup_bool = bool(tup)
print(f"Tuple: {tup}, Boolean: {tup_bool}")
# tuple to set conversion
tup_set = set(tup)
print(f"Tuple: {tup}, Set: {tup_set}")
# tuple to list conversion
tup_list = list(tup)
print(f"Tuple: {tup}, List: {tup_list}")
# tuple to dict conversion (using indices as keys)
tup_dict = { i: tup[i] for i in range(len(tup)) }
print(f"Tuple: {tup}, Dictionary: {tup_dict}")
```

```python
# set to str conversion
st = {7, 8, 9}
st_str = str(st)
print(f"Set: {st}, String: '{st_str}'")
# set to int conversion (will raise an error)
# st_int = int(st)  # Uncommenting this line will raise a TypeError
# set to float conversion (will raise an error)
# st_float = float(st)  # Uncommenting this line will raise a TypeError
# set to bool conversion
st_bool = bool(st)
print(f"Set: {st}, Boolean: {st_bool}")
# set to list conversion
st_list = list(st)
print(f"Set: {st}, List: {st_list}")
# set to tuple conversion
st_tuple = tuple(st)
print(f"Set: {st}, Tuple: {st_tuple}")
# set to dict conversion (using elements as keys with dummy values)
st_dict = { item: None for item in st }
print(f"Set: {st}, Dictionary: {st_dict}")
# set to list conversion
st_list = list(st)
print(f"Set: {st}, List: {st_list}")
```
