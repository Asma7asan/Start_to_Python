# A dictionary is a data structure that stores data in key–value pairs.
# - Key (like a name)
# - Value (the data you want to store)
# - Unordered and can not be sorted

```python
# dictionary example
dic = {}
dic['one'] = "one"
dic[2] = "two"
dic_1 ={'name':'python','age':20 , 'country':'china'}
print(dic)          # Create a dictionary and add key-value pairs to it, then print the dictionary.
print(dic[2])     # Access and print the value associated with the key 'name' in the second dictionary.
print(dic_1)        # Create another dictionary using a different method and print it.
print(dic.keys())   # Print all the keys in the first dictionary.
print(dic.values()) # Print all the values in the first dictionary.
```

```python
# keys : 
student = {
    'name': 'Ali',
    'age': 21,  
    'courses': ['Html', 'C++', 'Python']
}
print(student.keys())
```

```python
# values :
student = {
    'name': 'Ali',  
    'age': 21,
    'courses': ['Html', 'C++', 'Python']
}
print(student.values())
```

```python
# type : data type of the variable
dic = {'one': 1, 'two': 2, 'three': 3}
print("Data type of dic:", type(dic))
```

```python
# items :
student = { 
    'name': 'Ali',  
    'age': 21,
    'courses': ['Html', 'C++', 'Python']
}
print(student.items())
```

```python
# delete key-value pair :
student = {
    'name': 'Ali',  
    'age': 21,
    'courses': ['Html', 'C++', 'Python']
}
del student['age']
# del student['Ali']     # This will raise a KeyError since 'Ali' is not a key in the dictionary.
print(student)
```

```python
# clear dictionary :
student = { 
    'name': 'Ali',  
    'age': 21,
    'courses': ['Html', 'C++', 'Python']
}
student.clear()
print(student)  # This will print an empty dictionary: {
```

```python
# len() function to get the number of key-value pairs in a dictionary :
student = {
    'name': 'Ali',  
    'age': 21,
    'courses': ['Html', 'C++', 'Python']
}   
print(len(student))  # This will print 3
```

```python
# get() method to access values :
student = {
    'name': 'Ali',  
    'age': 21,
    'courses': ['Html', 'C++', 'Python']
}
print(student.get('name'))  # This will print 'Ali'
print(student.get('address', 'Not Found'))  # This will print 'Not Found' since 'address' key does not exist.
```

```python
# update() method to add key-value pairs :
student = {
    'name': 'Ali',  
    'age': 21,
    'courses': ['Html', 'C++', 'Python']
}   
student.update({'address': '123 Main St', 'age': 22})
print(student)  # This will print the updated dictionary with the new key-value pair and updated age.
```

```python
# copy() method to create a shallow copy of a dictionary :
student = {
    'name': 'Ali',  
    'age': 21,
    'courses': ['Html', 'C++', 'Python']
}
student_copy = student.copy()
print(student_copy)  # This will print the copied dictionary.
```

```python
# popitem() method to remove and return an arbitrary key-value pair :
student = {
    'name': 'Ali',  
    'age': 21,
    'courses': ['Html', 'C++', 'Python']
}
removed_item = student.popitem()
print(removed_item)  # This will print the removed key-value pair.
print(student)       # This will print the dictionary after removing the item.
```

```python
# setdefault() method to get the value of a key, and set it if the key does not exist :
student = {
    'name': 'Ali',  
    'age': 21,
    'courses': ['Html', 'C++', 'Python']
}
address = student.setdefault('address', 'Not Provided')
print(address)  # This will print 'Not Provided'
print(student)  # This will print the dictionary with the new 'address' key added.
```

```python
# items() method to get a view object of key-value pairs :
student = { 
    'name': 'Ali',  
    'age': 21,
    'courses': ['Html', 'C++', 'Python']
}
print(student.items())  # This will print a view object containing the key-value pairs of the dictionary.
```

```python
# fromkeys() method to create a new dictionary with specified keys and a default value :
new_dict = dict.fromkeys(['name', 'age', 'courses'], 'Unknown')
print(new_dict)  # This will print {'name': 'Unknown', 'age': 'Unknown', 'courses': 'Unknown'}
```
