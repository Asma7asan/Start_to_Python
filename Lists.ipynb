# A List is a data structure in Python used to store multiple items in one variable.
# - You can store numbers, strings, booleans, or even other lists.
# - Support Indexing and slicing
# - Can be nested
# - Mutable
# - item is not unique
# - Can have different data type

```python
# append
list_1 = [1, 2, 3, 4 ,5 ]   # a list data type integer
list_2 = ['a', 'b', 'c' ]    # a list data type string
list_3 = [True, False, True ]   # a list data type boolean
list_4 = [1.1, 2.2, 3.3 ]    # a list data type float
list_5 = []    # an empty list
list = [123 , 'abc', 45.67 , True ]    # a mixed data type 
print("Before append:", list)
list.append(list_1)
print("lists are mutable", list)
# lists are mutable, meaning you can change their content without changing their identity.
list.append(list_2)
print("After append:", list)
```

```python
#Lists
list_1 =[123 , 'Ali']
list = ['abc' , 123 , 45.67 , True , 100]    # a mixed data type
print("Befor :", list)
print(list[0])   # accessing the first element
print(list[1:3])   # accessing elements from index 1 to 2
print(list[-1])   # accessing the last element
print(list[2:])   # accessing elements from index 2 to the end
print(list * 2)   # repeating the list
print(list + list_1)   # concatenating lists
print(list[:])   # slicing the entire list
```

```python
# type : data type of the variable
list_1 = [1 , 'a' , 3.14 , True , 'hello']
print("Data type of list_1:", type(list_1))
```

```python
# len : length of list
list = ['a', 'b', 'c', 'd', 'e']
length = len(list)
print("Length of the list:", length)  # Output: 5
```

```python
# pop : remove and return element at index
list = ['a', 'b', 'c', 'd', 'e']
removed_element = list.pop(2)  # removes 'c'
print("Removed element:", removed_element)  # Output: 'c'
print("List after pop:", list)  # Output: ['a', 'b', 'd', 'e']
# * If no index is specified, pop() removes and returns the last item in the list.
```

```python
# remove : remove first occurrence of value
list = ['a', 'b', 'c', 'd', 'e', 'c']
list.remove('c')  # removes the first 'c'
print("List after remove:", list)  # Output: ['a', 'b', 'd', 'e', 'c']
# * If the value is not found, a ValueError is raised.
```

```python
# reverse : reverse the list in place
list = [1, 2, 3, 4, 5]
list.reverse()
print("Reversed list:", list)  # Output: [5, 4, 3, 2, 1]
# * The reverse() method modifies the original list and does not return a new list.
```

```python
# sort : sort the list in ascending order
list = [5, 2, 9, 1, 5, 6]
list.sort()
print("Sorted list:", list)  # Output: [1, 2, 5, 5, 6, 9]
# * The sort() method modifies the original list and does not return a new list.
```

```python
# max : find the maximum value in the list
list = [3, 1, 4, 1, 5, 9, 2, 6, 5]
maximum = max(list)
print("Maximum value in the list:", maximum)  # Output: 9
```

```python
# min : find the minimum value in the list
list = [3, 1, 4, 1, 5, 9, 2, 6, 5]
minimum = min(list)
print("Minimum value in the list:", minimum)  # Output: 1
```

```python
# sum : calculate the sum of numeric elements in the list
list = [1, 2, 3, 4, 5 ]
total = sum(list)
print("Sum of the list elements:", total)  # Output: 15
```

```python
# count : count occurrences of a value in the list
list = ['a', 'b', 'c', 'a', 'd', 'a']
count_a = list.count('a')
print("Count of 'a' in the list:", count_a)  # Output: 3
```

```python
# copy : create a shallow copy of the list
list = ['a', 'b', 'c', 'd', 'e']
copied_list = list.copy()
print("Original list:", list)  # Output: ['a', 'b', 'c
print("Copied list:", copied_list)  # Output: ['a', 'b', 'c', 'd', 'e']
# * Modifications to the copied list do not affect the original list.
```

```python
# index : find the index of the first occurrence of a value
list = ['a', 'b', 'c', 'd', 'e']
index_c = list.index('c')
print("Index of 'c' in the list:", index_c)  # Output: 2
# * If the value is not found, a ValueError is raised.
```

```python
# insert : insert an element at a specific index
list = ['a', 'b', 'd', 'e']
list.insert(2, 'c')  # inserts 'c' at index 2
print("List after insert:", list)  # Output: ['a', 'b', '
```

```python
# extend : extend list by appending elements from another iterable
list = ['a', 'b', 'c']
list.extend(['d', 'e', 'f'])
print("List after extend:", list)  # Output: ['a', 'b', 'c', 'd', 'e', 'f']
```

```python
# clear : remove all items from the list
list = ['a', 'b', 'c', 'd', 'e']
list.clear()
print("List after clear:", list)  # Output: []
```
