# A Tuples is a data type in Python similar to a list, but with one key difference:
# - you cannot change, add, or remove elements after itâ€™s created.
# - Ordered
# - Allow multiple data types
# - Immutable (cannot be changed)
# - Faster than lists

```python
# append : Not applicable for tuples as they are immutable
# remove : Not applicable for tuples as they are immutable
# pop : Not applicable for tuples as they are immutable
# clear : Not applicable for tuples as they are immutable
# sort : Not applicable for tuples as they are immutable
# reverse : Not applicable for tuples as they are immutable
# copy : Not applicable for tuples as they are immutable
# extend : Not applicable for tuples as they are immutable
# insert : Not applicable for tuples as they are immutable
```

```python
# difference between lists and tuples :
# - Lists are mutable, tuples are immutable
# - Lists use square brackets [], tuples use parentheses ()
# - Lists can be changed (add, remove, modify elements), tuples cannot be changed after creation
# - Tuples are typically used for fixed collections of items, while lists are used for collections that may change.
# - Tuples are immutable sequences, typically used to store collections of heterogeneous data.
tuple_example = (10, 20, 30)
list_example = [10, 20, 30]
print("Tuple example:", tuple_example)
print("List example:", list_example)
# tuple_example[0] = 100  # This will raise an error
list_example[0] = 100  # This will work
print("Modified list example:", list_example)

```

```python
# Tuples
tuple_1 = (1 , 'a' , 3.14 , True , 'hello')
tuple_2 = 2 , 'b' , 2.71 , False , 'world'  # Parentheses are optional
tuple_3 = tuple_1 + ('new', 42) + tuple_2 # Concatenation
print("Tuple:", tuple_1)
print("First element of tuple:", tuple_1[0])   
print("Last element of tuple:", tuple_1[-1])
print("Slice of tuple (1:4):", tuple_1[1:4])
print(tuple_1[2:])
print(tuple_1 * 2)  # Repetition
print(tuple_1 + tuple_2)  # Concatenation
print(tuple_3)
```

```python
# len : length of tuple
tuple_1 = (1 , 'a' , 3.14 , True , 'hello')
print("Length of tuple_1:", len(tuple_1))   
```

```python
# type : data type of the variable
tuple_1 = (1 , 'a' , 3.14 , True , 'hello')
print("Data type of tuple_1:", type(tuple_1))
```

```python
# index :use index to find the position of an element and enter the value to find its index
tuple_1 = (1 , 'a' , 3.14 , True , 'hello')
print("Index of 'a' in tuple_1:", tuple_1.index('a'))
print("Index of 3.14 in tuple_1:", tuple_1.index(3.14))
```

```python
# count : count occurrences of an element in the tuple
tuple_1 = (1 , 'a' , 3.14 , True , 'hello' , 'a' , 1)
print("Count of 'a' in tuple_1:", tuple_1.count('a'))
print("Count of 1 in tuple_1:", tuple_1.count(1))
```

```python
# immutability of tuples - trying to change an element will raise an error and cannot add or remove elements
tuple_1 = (1 , 'a' , 3.14 , True , 'hello')
# tuple_1[0] = 100              # This will raise an error
# tuple_1.append('new')         # This will raise an error 
print("Original tuple_1:", tuple_1)
# To modify a tuple, you need to create a new one
tuple_2 = tuple_1 + ('new', 42)
print("Modified tuple_2:", tuple_2)
```

```python
# del : delete entire tuple
tuple_1 = (1 , 'a' , 3.14 , True , 'hello')
print("Original tuple_1:", tuple_1)
del tuple_1
#print("After deletion, tuple_1:", tuple_1)  # This will raise an error since tuple_1 is deleted
```

```python
# max : maximum value in a tuple of numbers
tuple_numbers = (10, 20, 5, 30, 15)
print("Maximum value in tuple_numbers:", max(tuple_numbers))
```

```python
# min : minimum value in a tuple of numbers
tuple_numbers = (10, 20, 5, 30, 15)
print("Minimum value in tuple_numbers:", min(tuple_numbers))
```
