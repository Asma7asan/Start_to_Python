# A while loop repeats a block of code as long as a condition is True.

```python
# Count from 1 to 5
i = 1
while i <= 5:
    print(i)
    i += 1  
```

```python
# Sum numbers until 10
total = 0
i = 1
while i <= 10:
    total += i
    i += 1

print("Sum:", total)
```

```python
# Print all even numbers from 2 to 20
i = 2
while i <= 20:
    print(i)
    i += 2
```

```python
# Print stars in a triangle using a while loop
i = 1
rows = 5

while i <= rows:
    print("*" * i)
    i += 1
```

```python
# Diamond
rows = 3
i = 1

# Top half
while i <= rows:
    print(" " * (rows - i) + "*" * (2 * i - 1))
    i += 1

# Bottom half
i = rows - 1
while i > 0:
    print(" " * (rows - i) + "*" * (2 * i - 1))
    i -= 1
```
