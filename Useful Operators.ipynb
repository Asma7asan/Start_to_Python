```python
# enumerate() : allows you to loop through counter to any iterable (list, tuple, string…).and get both: the index and the value of each item
fruits = ["apple", "banana", "orange"]
enumerate_fruits = enumerate(fruits)
print(list(enumerate_fruits))
```

```python
# in operator : use to quickly check if an object is in a list
List = ['a','b','c','d','e']
'c' in List
```

```python
# not in : use to quickly check if some object or variable is not present in a list
List = ['a','b','c','d','e']
'c' not in List
```

```python
# min : # use to quickly check the minimum of a list
List = [10, 20,30, 40, 50]
min(List)
```

```python
# max : # use to quickly check the maximum of a list
List = [10, 20,30, 40, 50]
max(List)
```
