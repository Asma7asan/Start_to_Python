# If statements : The if statement is used when you want your program to make a decision

# syntax
# if some_condition:
#    # execute some code
# elif some_other_condition:
#    # do something different
# else:
#    # do something else

```python
# example
grade = 85
if grade >= 90:
    print("Excellent")
elif grade >= 70:
    print("Good")
else:
    print("Needs improvement")
```

```python
# Checking if a number is positive
number = 78
if number > 0:
  print("The number is positive")
```

```python
# Checking if a number is negative
number = -7
if number < 0:
  print("The number is negative")
```

```python
# multiple conditions "Grade"
score = 75
if score >= 90:
  print("Grade: A")
elif score >= 80:
  print("Grade: B")
elif score >= 70:
  print("Grade: C")
elif score >= 60:
  print("Grade: D")
```

```python
# check value of the variable letter using else
letter = "A"
if letter == "B":
    print("letter is B")
else:
    if letter == "C":
        print("letter is C")
    else:
        if letter == "A":
            print("letter is A")
        else:
            print("letter isn't A, B and C")
```

```python
# check value of the variable letter using elif
letter = "A"
if letter == "B":
    print("letter is B")
elif letter == "C":
    print("letter is C")
elif letter == "A":
    print("letter is A")
else:
    print("letter isn't A, B and C")
```

```python
# Using elif Inside Nested if Statements
x = 100
y = 90
if x > 50:
    if y > 50:
        print("x is greater than 50")
    elif y==50:
         print("x is greater than 50 and y is 50")
    else:
        print("x is greater than 50 and y is less than 50")
```

```python
# even and odd number using if else
num = 7
if num % 2 == 0 :
  print("7 is even")
else:
  print("7 is odd")
```

```python
# conditions more complex 
a = 10
b = 20
c = 30
if (a + b) / c == 1 and c - b - a == 0:
    print('yes')
else:
    print('no')
```
