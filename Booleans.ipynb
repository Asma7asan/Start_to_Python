# A boolean (bool) is a data type in Python that has only two possible values: True or  False


```python
# True Values
print(bool('Python'))
print(bool(100))
print(bool(78.09))
print(bool(True))
print(bool([1, 2, 3]))
print(bool((1, 2, 3)))
print(bool({'name': 'Alice'}))

```

```python
# False Values
print(bool(''))
print(bool(""))
print(bool(0))
print(bool(False))
print(bool([]))
print(bool(()))
print(bool({}))
print(bool(None))
print(bool(0.0))
print(bool(range(0)))
print(bool(set()))

```
